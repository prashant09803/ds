```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Iris.csv')
```

```python
df
```

```python
x = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=40)
x_test
```

```python
from sklearn.preprocessing import LabelEncoder
la = LabelEncoder()
y = la.fit_transform(y)
y
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
```

```python
y_predicted = model.predict(x_test)
y_predicted
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm =confusion_matrix(y_test,y_predicted)
cm
```

```python
cl_report = classification_report(y_test,y_predicted)
print(cl_report)
```

```python
cm_df = pd.DataFrame(cm,
index =['Iris-setosa','Iris-versicolor','Iris-virginica'],
columns =['Iris-setosa','Iris-versicolor','Iris-virginica'])
cm_df
```

```python
import seaborn as sns
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values')
plt.show()
```

```python
def accuracy_cm(tp,fn,fp,tn):
    return (tp+tn)/(tp+fp+tn+fn)
    
def precision_cm(tp,fp,tn,fn):
    return tp/(tp+fp)
    
def recall_cm(tp,fp,tn,fn):
    return tp/(tp+fn)
    
def error_cm(tp,fp,tn,fn):
    return 1-accuracy_cm(tp,fn,fp,tn)
```

```python
tp = cm[2][2]
fn = cm[2][0]+cm[2][1]
fp = cm[0][2]+cm[1][2]
tn = cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1]

print("Accuracy:",accuracy_cm(tp,fn,fp,tn))
print("Precision:",precision_cm(tp,fp,tn,fn))
print("Recall:",recall_cm(tp,fp,tn,fn))
print("Error:",error_cm(tp,fp,tn,fn))
```
