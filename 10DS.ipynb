```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("./Iris.csv")
df
```

```python
sns.histplot(x=df["SepalLengthCm"], kde=True)
plt.title("Histogram of Sepal Length")
```

```python
sns.histplot(x=df["SepalWidthCm"], kde=True)
plt.title("Histogram of Sepal Width")
```

```python
sns.histplot(x=df["PetalLengthCm"], kde=True)
plt.title("Histogram of Petal Length")
```

```python
sns.histplot(x=df["PetalWidthCm"], kde=True)
plt.title("Histogram of Petal Width")
```

```python
sns.boxplot(x="SepalLengthCm", y="Species", data=df)
plt.title("Boxplot for each feature in the dataset")
```
