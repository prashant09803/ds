```python
df = df.replace('?', np.nan)
df
```

```python
df.astype
```

```python
df.dtypes
```

```python
avg_crim = df['CRIM'].astype('float').mean()
avg_crim
```

```python
df["CRIM"].replace(np.nan, avg_crim)
df
```

```python
avg_zn = df["ZN"].astype("float").mean(axis=0)
print("Average of ZN:", avg_zn)
```

```python
df["ZN"].replace(np.nan, avg_zn, inplace=True)
df
```

```python
avg_indus = df["INDUS"].astype("float").mean(axis=0)
print("Average of INDUS:", avg_indus)
```

```python
df["INDUS"].replace(np.nan, avg_indus, inplace=True)
df
```

```python
df.dropna(inplace =True)
df
```

```python
df.isnull().sum()
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.displot(df['MEDV'], bins=30)
plt.show()
```

```python
correlation_matrix = df.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
X = df[['LSTAT' , 'RM']]
Y = df[['MEDV']]
X.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# splitting the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=40)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

```python
Y_predict = linear_model.predict(X_test)
```

```python
r2_score(Y_test, Y_predict)
```

```python
mean_squared_error(Y_test, Y_predict)
```
